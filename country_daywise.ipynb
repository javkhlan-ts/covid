```python
# import dataframes from url
confirmed = pd.read_csv(urlConfirmed)
death = pd.read_csv(urlDeath)
recovered = pd.read_csv(urlRecovered)
```

```python
#remove unnecessary columns
confirmed = confirmed.drop(['Province/State', 'Lat', 'Long'], axis=1)
death = death.drop(['Province/State', 'Lat', 'Long'], axis=1)
recovered = recovered.drop(['Province/State', 'Lat', 'Long'], axis=1)
```

```python
confirmed
```

```python
#rename columns
confirmed = confirmed.rename(columns={"Country/Region": "Country"})
death = death.rename(columns={"Country/Region": "Country"})
recovered = recovered.rename(columns={"Country/Region": "Country"})
```

```python
confirmed.head()
```

```python
#some countries have multiple province entries, hence group them
confirmed = confirmed.groupby('Country').sum().reset_index()
death = death.groupby('Country').sum().reset_index()
recovered = recovered.groupby('Country').sum().reset_index()
```

```python
confirmed.head(3)
```

```python
# melt all dates into columns
confirmed = pd.melt(confirmed, id_vars=['Country'])
death = pd.melt(death, id_vars=['Country'])
recovered = pd.melt(recovered, id_vars=['Country'])
```

```python
confirmed
```

```python
#rename columns
confirmed = confirmed.rename(columns={"variable": "Date", "value": "Confirmed"})
death = death.rename(columns={"variable": "Date", "value": "Death"})
recovered = recovered.rename(columns={"variable": "Date", "value": "Recovered"})
```

```python
#merge operation, not assigned, temp
pd.merge(confirmed, death)
```

```python
#merge operation, not assigned, temp
pd.merge(pd.merge(confirmed, death), recovered)
```

```python
#merge operation
df = pd.merge(pd.merge(confirmed, death), recovered)
```

```python
df.head()
```

```python
#rearrange columns
df=df.reindex(columns=['Date', 'Country', 'Confirmed', 'Death', 'Recovered'])
```

```python
df.head()
```

```python
df['Active'] = df['Confirmed'] - df['Death'] - df['Recovered']
```

```python
df.tail()
```

```python
#Date string to Data object
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#New cases
temp = df.groupby(['Country', 'Date'])['Confirmed', 'Death', 'Recovered']
temp = temp.sum().diff().reset_index()
mask = temp['Country'] != temp['Country'].shift(1)
temp.loc[mask, 'Confirmed'] = np.nan
temp.loc[mask, 'Death'] = np.nan
temp.loc[mask, 'Recovered'] = np.nan

# renaming columns
temp.columns = ['Country', 'Date', 'New cases', 'New death', 'New recovered']

# merging new values
df = pd.merge(df, temp, on=['Country', 'Date'])

# filling na with 0
df = df.fillna(0)

# fixing data types
cols = ['New cases', 'New death', 'New recovered']
df[cols] = df[cols].astype('int')

# 
df['New cases'] = df['New cases'].apply(lambda x: 0 if x<0 else x)
```
